```python
import pandas as pd
import numpy
```

```python
# 파일 불러와서 변수에 저장
placeinfo = pd.read_csv('대여소.csv', encoding='cp949')
loginfo = pd.read_excel('log.xlsx')
```

```python
placeinfo
```

```python
# 대여소번호에 NA값 있는 행 삭제
placeinfo = placeinfo.dropna(subset=['대여소\n번호'])
```

```python
placeinfo
```

```python
#컬럼명 깔끔히 정렬
placeinfo.columns = ['대여소번호', '대여소명', '소재지', '주소', '위도', '경도', '설치시기', 'lcd', 'qr', '운영방식']
```

```python
placeinfo
```

```python
# 대여소번호가 부동소수점이라서 int 변경후 str로 재변경
placeinfo = placeinfo.astype({'대여소번호' : 'int'})
placeinfo = placeinfo.astype({'대여소번호' : 'str'})
```

```python
placeinfo
```

```python
loginfo
```

```python
#loginfo 컬럼명 깔끔히 정렬
loginfo.columns = ['구', '대여소명', '반납일', '건수']
```

```python
# loginfo 대여소명 컬럼에서 대여소번호 찾아서 새로운 칼럼 만들어주기
loginfo['대여소번호'] = loginfo.대여소명.str.split('.').str[0]

```

 loginfo['대여소번호'] = loginfo.대여소명.str[:loginfo.대여소명.str.index('.')]
 #새로운 방법

```python
loginfo
```

```python
#loginfo에 대여소번호칼럼 값 중 '대여소명 없음' 삭제 (굳이해줄필요없었음)
abc = loginfo[loginfo['대여소번호'] == '대여소명 없음'].index
loginfo = loginfo.drop(abc)
```

```python
loginfo
```

```python
loginfo.info()
```

```python
# 조인으로 덮어쓰기
total_info= pd.merge(loginfo, placeinfo, left_on='대여소번호', right_on='대여소번호', how='left')
print(total_info)
```

```python
#csv파일로 내보내기
total_info.to_csv("total_Csv")
```
